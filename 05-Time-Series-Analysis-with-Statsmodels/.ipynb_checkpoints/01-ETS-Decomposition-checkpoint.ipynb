## ETS

ETS stands for Error/Trend/Seasonality Models.

ETS models will take each of those terms for "smoothing" and may add them, multiply them or even just leave some of them out.

## Error/Trend/Seasonality Models
As we begin working with <em>endogenous</em> data ("endog" for short) and start to develop forecasting models, it helps to identify and isolate factors working within the system that influence behavior. Here the name "endogenous" considers internal factors, while "exogenous" would relate to external forces. These fall under the category of <em>state space models</em>, and include <em>decomposition</em> (described below), and <em>exponential smoothing</em> (described in an upcoming section).

The <a href='https://en.wikipedia.org/wiki/Decomposition_of_time_series'>decomposition</a> of a time series attempts to isolate individual components such as <em>error</em>, <em>trend</em>, and <em>seasonality</em> (ETS). We've already seen a simplistic example of this in the <strong>Introduction to Statsmodels</strong> section with the Hodrick-Prescott filter. There we separated data into a trendline and a cyclical feature that mapped observed data back to the trend.

## Seasonal Decomposition
Statsmodels provides a <em>seasonal decomposition</em> tool we can use to separate out the different components. This lets us see quickly and visually what each component contributes to the overall behavior.


We apply an <strong>additive</strong> model when it seems that the trend is more linear and the seasonality and trend components seem to be constant over time (e.g. every year we add 10,000 passengers).<br>
A <strong>multiplicative</strong> model is more appropriate when we are increasing (or decreasing) at a non-linear rate (e.g. each year we double the amount of passengers).

For these examples we'll use the International Airline Passengers dataset, which gives monthly totals in thousands from January 1949 to December 1960.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(rc = {'figure.figsize':(11,7)})
```

```python
airline = pd.read_csv('../Data/airline_passengers.csv',index_col='Month',parse_dates=True)
```

```python
airline.dropna(inplace=True)
```

```python
airline.head()
```

```python
airline.plot();
```

The dataset shows thousands of passengers flying per month from the year 1949 to the year 1960. 


On looking at the plot, we can see the following:  
We can see a general trend upward. And the trend can be of two different types - linear or exponential trend. 
And it is hard to tell which trend it is. And as it is hard to tell from the raw dataset so it will be nice to seperate out the 
trend component. The reason it is hard to tell because there is seasonality to the dataset (Peak travels happen during the summer month). 

Based on this chart, it looks like the trend in the earlier days is increasing at a higher rate than just linear (although it is a  bit hard to tell from this one plot).

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(airline['Thousands of Passengers'], model='multiplicative')  # model='mul' also works
result.plot();
```

ETS decomposition results/returns four plots and all four of them are aligned along the same x-axis' timestamp.

__Trend plot__ shows the general growth/decline pattern of the actual dataset(plotted earlier).

__Isolated seasonal component:__ Seasonal component obtained from the actual dataset and it is perfectly repeating and that allows us to analyze the component based on seasonality.

__Residual component:__ is for things not explained by trend or seasonality terms. And thats why it almost looks like noise but it can also be a good indicator of where there is noise in the dataset and where the dataset tends to be cleaner.

### Important pts to note before applying ETS decomposition:
There are 2 types of ETS models:
1. Additive model
2. Multiplicative model


If we observe the trend we see the trend increases slightly higher than linear so use multiplicative model here.


On typing result by itself we get a Decomposeresult object. 
`result` - `<statsmodels.tsa.seasonal.DecomposeResult at 0x1d63de72668>`

To increase the size of the resulting plot can use the following codes:
`sns.set(rc={'figure.figsize':(10,8)})`

Can also plot the plots independently.
Eg: `result.trend.plot()`

### Deseasonalization and seasonally adjusted values
The seasonality factor makes it difficult for us to identify whether the data is depicting an upward trend or a downward trend. Therefore, we need to remove the seasonality factor from our data. The process of removing the seasonality factor from the data is known as __deseasonalization__ and the resultant values are known as __seasonally adjusted values__.

Let's calculate the seasonally adjusted values in the AirPassengers data using the below formula:

Seasonally adjusted values = Time Series Data - (Seasonal) = (Trends) + (Random)
