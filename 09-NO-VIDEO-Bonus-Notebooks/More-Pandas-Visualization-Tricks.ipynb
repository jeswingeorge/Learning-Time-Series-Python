# Additional Pandas Visualization Tricks

```python
# RUN THIS CELL
import pandas as pd
%matplotlib inline

df2 = pd.read_csv('../03-Pandas-Visualization/df2.csv')
```

## Quickly add one axis label
We mentioned that although we can pass a title into the pandas .plot() function, we can't pass axis labels this way.<br>
However, we <em>can</em> quickly add a label to either the x- or y-axis by running a <tt>.set_xlabel()</tt> or <tt>.set_ylabel()</tt> method:

```python
df2.plot(figsize=(8,3), title='Custom Pandas Plot').set_xlabel('X-axis data');
```

Unfortunately, we can't add both labels this way:

```python
df2.plot(figsize=(8,3)).set_xlabel('X-axis data').set_ylabel('Y-axis data');
```

## Remove unwanted axis labels
Sometimes pandas automatically places the column name as an x or y axis label. To remove them we can call the label on the axis object and set_visible to False. Alternatively, we can simply pass in empty strings as labels.

```python
df2.plot.scatter(x='a',y='b',figsize=(4,2));
```

```python
ax = df2.plot.scatter(x='a',y='b',figsize=(4,2))
ax.xaxis.get_label().set_visible(False)
ax.yaxis.get_label().set_visible(False)
```

```python
# ALTERNATIVELY:
ax = df2.plot.scatter(x='a',y='b',figsize=(4,2))
ax.set(xlabel='', ylabel='');
```

## Adding commas to y-axis values

```python
import matplotlib.ticker as ticker
formatter = ticker.StrMethodFormatter('{x:,.0f}')

df=pd.DataFrame([1000000,800000,60000,200,2000],index=range(5))

ax = df.plot(figsize=(4,2),legend=False)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel='', ylabel='')
ax.yaxis.set_major_formatter(formatter);
```

## Adding dollar signs to y-axis values

```python
import matplotlib.ticker as ticker
formatter = ticker.FormatStrFormatter('$%1.2f')

df=pd.DataFrame([10,3.5,.4,7.5,9.1],index=range(5))

ax = df.plot(figsize=(4,2),legend=False)
ax.autoscale(axis='x',tight=True)
ax.set(xlabel='', ylabel='price')
ax.yaxis.set_major_formatter(formatter);
```
